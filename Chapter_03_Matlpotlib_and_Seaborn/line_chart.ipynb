```python
import matplotlib.pyplot as plt
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years,runs)
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Sachin Tendulkar Runs")
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years , sachin , "D--g" , label = "Sachin")  # first shape of points then type of line and then color of line 
plt.plot(years , kohli , "b-*" ,label = "Kohli")      # we can also do first color then type of line then shape of points 
plt.plot(years,sehwag , linewidth = 3 , label = "sehwag")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Comparison between runs of kohli , sehwag and sachin")
plt.grid()
plt.legend()
```
