```python
import matplotlib.pyplot as plt
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ["red","pink","green","yellow"]  # we can also use hex codes here
explode = [0,0,0,0.1]
plt.pie(
    runs ,
    labels = labels ,  # colors and labels accept a list
    colors = colors ,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'} ,  # wedgeprops to add styling in your chart(check on mdn)
    explode = explode,  # explode to make some(usually 1) pie outer than others
    shadow = True,  # add shadows for 3-D feel
    startangle = 70,  # change angle of pie chart
    autopct = "%.1f%%"  # to check % share of everyone 
    # f => all decimal places , .f => no decimal , .1f=> upto 1 decimal place , .2f => upto 2 decimal places and so on......
)  
plt.title("Career Runs")
```
